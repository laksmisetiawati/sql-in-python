```python
# Load libraries
import pandas as pd

from google.cloud import bigquery

print("success load libraries!")
```

```python
# Google Cloud project id
project_id = "kumpulan-dataset-ls"

# Create a bigquery client
client = bigquery.Client(project = project_id)
```

```python
# Run query
q_products = """
          SELECT * FROM `kumpulan-dataset-ls.ecommerce_b2b.products`
        """
product_query_job = client.query(q_products)

# Load query return as dataframe
df_products = product_query_job.to_dataframe()
df_products.head(3)
```
